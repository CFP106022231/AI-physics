# 1. Data structure

```python
npz = np.load('/kaggle/input/phys591000-electron-shower/emulsion.npz', allow_pickle=True)
df= pd.DataFrame.from_dict({item: npz[item] for item in npz.files}, orient='index')
print(df.shape)
print(npz.files)
df
```

```python
truth_energy_0 = npz["truth_energy"][npz["label"]==0]
truth_energy_1 = npz["truth_energy"][npz["label"]==1]

num_bt_of_shower_max_0 = npz["num_bt_of_shower_max"][npz["label"]==0]
num_bt_of_shower_max_1 = npz["num_bt_of_shower_max"][npz["label"]==1]

num_total_basetracks_0 = npz["num_total_basetracks"][npz["label"]==0]
num_total_basetracks_1 = npz["num_total_basetracks"][npz["label"]==1]

layer_index_of_shower_max_0 = npz["layer_index_of_shower_max"][npz["label"]==0]
layer_index_of_shower_max_1 = npz["layer_index_of_shower_max"][npz["label"]==1]

basetracks_0 = npz["basetracks"][npz["label"]==0]
basetracks_1 = npz["basetracks"][npz["label"]==1]
```

# 2. Data visualization

```python
# fig, ax = plt.subplots(1,1,figsize=(6,6))
# plt.rcParams["figure.figsize"] = (15,8)

feature = [truth_energy_0,num_bt_of_shower_max_0,num_total_basetracks_0,layer_index_of_shower_max_0]
name = ["truth_energy_0","num_bt_of_shower_max_0","num_total_basetracks_0","layer_index_of_shower_max_0"]

list = [0,1,2,3]


for i in range(3):
    fig,ax =  plt.subplots(1,3,figsize=(15,4))
    #print("i="+str(i))
    list.remove(list[i])
    m=0
    for j in list:
        #print(j)
        ax[m].scatter(feature[i],feature[j], c= "blue")
        ax[m].tick_params(axis='x', labelsize=10)
        ax[m].tick_params(axis='y', labelsize=10)
        ax[m].set_xlabel("\n"+name[i], fontsize=15,horizontalalignment='right',x=1)
        ax[m].set_ylabel(name[j], fontsize=15,horizontalalignment='right',y=1)
        list = [0,1,2,3]
        m+=1
        
#################################################################################################################
feature_1 = [truth_energy_1,num_bt_of_shower_max_1,num_total_basetracks_1,layer_index_of_shower_max_1]
name_1 = ["truth_energy_1","num_bt_of_shower_max_1","num_total_basetracks_1","layer_index_of_shower_max_1"]

list = [0,1,2,3]

for i in range(3):
    fig,ax =  plt.subplots(1,3,figsize=(15,4))
    #print("i="+str(i))
    list.remove(list[i])
    m=0
    for j in list:
        #print(j)
        ax[m].scatter(feature_1[i],feature_1[j], c= "red")
        ax[m].tick_params(axis='x', labelsize=10)
        ax[m].tick_params(axis='y', labelsize=10)
        ax[m].set_xlabel("\n"+name_1[i], fontsize=15,horizontalalignment='right',x=1)
        ax[m].set_ylabel(name_1[j], fontsize=15,horizontalalignment='right',y=1)
        list = [0,1,2,3]
        m+=1
```

```python
plt.rcParams["figure.figsize"] = (15,8)
fig,ax =  plt.subplots(2,2)

ax[0,0].hist(truth_energy_0,bins=100,color='r',alpha=0.5,label='electrons');
ax[0,0].hist(truth_energy_1,bins=100,color='b',alpha=0.5,label='muons');
ax[0,0].legend()
ax[0,0].set_title('truth_energy',fontsize=20)

ax[0,1].hist(num_bt_of_shower_max_0,bins=50,color='r',alpha=0.5,label='electrons');
ax[0,1].hist(num_bt_of_shower_max_1,bins=50,color='b',alpha=0.5,label='muons');
ax[0,1].legend()
ax[0,1].set_title('num_bt_of_shower_max',fontsize=20)

ax[1,0].hist(num_total_basetracks_0,bins=50,color='r',alpha=0.5,label='electrons');
ax[1,0].hist(num_total_basetracks_1,bins=50,color='b',alpha=0.5,label='muons');
ax[1,0].legend()
ax[1,0].set_title('num_total_basetracks',fontsize=20)

ax[1,1].hist(layer_index_of_shower_max_0,bins=50,color='r',alpha=0.5,label='electrons');
ax[1,1].hist(layer_index_of_shower_max_1,bins=50,color='b',alpha=0.53,label='muons');
ax[1,1].legend()
ax[1,1].set_title('layer_index_of_shower_max',fontsize=20)
```
