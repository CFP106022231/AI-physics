# Data structure

```python
mytestfile = "../input/phys591000-top-jets/fixed/processed-pythia82-lhc13-all-pt1-50k-r1_h022_e0175_t220_nonu_withPars_truth_6.z"
f = h5py.File(mytestfile, 'r')
list(f.keys())
```

```python
dset = f['t_allpar_new']
dset.shape
```

```python
data = np.array(f['t_allpar_new'])
data = pd.DataFrame(data)
data.head()
```

```python
data.columns
```

```python
print('data.shape : ',data.shape )
```

```python
# classify data
j0 = data[data['j_g']==1]
j1 = data[data['j_q']==1]
j2 = data[data['j_w']==1]
j3 = data[data['j_z']==1]
j4 = data[data['j_t']==1]
jet = ['j_g', 'j_q','j_w', 'j_z', 'j_t']
for i in range(5):
    print('number of',jet[i],' is ',data[data[jet[i]]==1].shape[0])
```

# Data Visualization

```python
jet = ['j_g', 'j_q','j_w', 'j_z', 'j_t']

num = -1
for j in range(7):
    fe = data.columns[j*10:(j+1)*10]
    b = 200
    fig , ax = plt.subplots(figsize=(30,8))
    fig.subplots_adjust(hspace=0.3, wspace=0.3)
    for i in range(len(fe)):
        num+=1
        plt.subplot(2, 5, i+1)
        plt.hist(j0[fe[i]], histtype='step', linestyle='-', bins=b, label= jet[0])
        plt.hist(j1[fe[i]], histtype='step', linestyle='-', bins=b, label= jet[1])
        plt.hist(j2[fe[i]], histtype='step', linestyle='-', bins=b, label= jet[2])
        plt.hist(j3[fe[i]], histtype='step', linestyle='-', bins=b, label= jet[3])
        plt.hist(j4[fe[i]], histtype='step', linestyle='-', bins=b, label= jet[4])
        plt.title(str(num)+'  '+fe[i],size=20)
    plt.legend(bbox_to_anchor=(1.05, 1.2), loc='center left',fontsize='xx-large')
```
