### mean of W jet

```python
import math
ms = 0
for j in range(0,14):
    for i in range(0,10037):
        ms = ms + (W_jet.iloc[i,3+j])**2
    ms = ms/10037  
    rms = math.sqrt(ms)
    print(rms)
```

### mean of QCD jet

```python
ms = 0
for j in range(0,14):
    for i in range(0,9963):
        ms = ms + (QCD_jet.iloc[i,3+j])**2
    ms = ms/9963 
    rms = math.sqrt(ms)
    print(rms)
```

# 3. Graphing data (1D)

a. 1D plot fo each feature : assign xmin, xmax, binsize

b. superpose and compare two historams, one in red and the other in blue

### 3.(a) 1D plot for each feature of W jet

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# a. 1D plot fo each feature : assign xmin, xmax, binsize 
for i in range(3,18):
    sns.distplot(a=W_jet.iloc[:,i], kde=False)
    min_tmp = np.min(W_jet.iloc[:,i])
    max_tmp = np.max(W_jet.iloc[:,i])
    plt.xlim(min_tmp,max_tmp)
    plt.show()

# Hint: https://www.kaggle.com/learn/data-visualization  Lesson:5
```

### 3.(a)1D plot for each feature of QCD jet

```python
for i in range(3,18):
    sns.distplot(a=QCD_jet.iloc[:,i], kde=False)
    min_tmp = np.min(QCD_jet.iloc[:,i])
    max_tmp = np.max(QCD_jet.iloc[:,i])
    plt.xlim(min_tmp,max_tmp)
    plt.show()
```

### 3.(b) superpose
superpose and compare two historams, one in red and the other in blue
Histograms for each species

```python

for i in range(3,18):
    
    sns.distplot(a=W_jet.iloc[:,i], label="W_jet", kde=False, norm_hist =0, color = "red" )
    sns.distplot(a=QCD_jet.iloc[:,i], label="QCD_jet", kde=False, norm_hist =0, color = "blue" )

    
    
    min_tmp = min(np.min(W_jet.iloc[:,i]),np.min(QCD_jet.iloc[:,i])) 
    max_tmp = max(np.max(W_jet.iloc[:,i]),np.max(QCD_jet.iloc[:,i])) 
    if i ==7:
        max_tmp = 200
    if i ==6:
        max_tmp = 30
    
    plt.xlim(min_tmp,max_tmp)

    plt.legend()
    plt.show()
    # Hint: https://www.kaggle.com/learn/data-visualization  Lesson:5
    
    #Feedback: I think there problems with these data: D21, D22, C21, C22
```
